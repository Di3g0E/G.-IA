```python
from collections import deque

def bfs(graph, start):
    visited = set()
    queue = deque([start])

    while queue:
        current_node = queue.popleft()
        if current_node not in visited:
            print(current_node, end=' ')
            visited.add(current_node)
            queue.extend(neighbor for neighbor in graph[current_node] if neighbor not in visited)

# Ejemplo de uso
# Se representa un grafo como un diccionario de listas de adyacencia
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F', 'G'],
    'D': ['B'],
    'E': ['B', 'H'],
    'F': ['C'],
    'G': ['C'],
    'H': ['E']
}

# Llamada a la función BFS desde el nodo 'A'
print("Recorrido BFS:")
bfs(graph, 'A')

# En este ejemplo, la función bfs realiza la búsqueda en amplitud utilizando una cola (deque).
# Se inicia desde un nodo de inicio ('A') y continúa visitando nodos en niveles sucesivos.
# La cola se utiliza para mantener un seguimiento de los nodos a visitar en el próximo nivel.

# Puedes ajustar el grafo y el nodo de inicio según tus necesidades.
# Similar a la búsqueda en profundidad (DFS), esta implementación asume que el grafo es no dirigido y no tiene ciclos infinitos.
# Para grafos más grandes o complejos, también puede ser necesario implementar manejo de ciclos y otros elementos de control adicionales.
```
