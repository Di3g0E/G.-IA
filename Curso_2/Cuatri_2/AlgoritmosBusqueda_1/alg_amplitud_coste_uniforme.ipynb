```python
import heapq

def uniform_cost_search(graph, start, goal):
    visited = set()
    priority_queue = [(0, start)]  # Cola de prioridad (costo, nodo)

    while priority_queue:
        current_cost, current_node = heapq.heappop(priority_queue)

        if current_node not in visited:
            print(current_node, end=' ')
            visited.add(current_node)

            if current_node == goal:
                print("\n¡Objetivo alcanzado!")
                return  # Se alcanzó el objetivo

            # Expandir nodos vecinos con su respectivo costo
            for neighbor, cost in graph[current_node]:
                if neighbor not in visited:
                    heapq.heappush(priority_queue, (current_cost + cost, neighbor))

# Ejemplo de uso
# Se representa un grafo como un diccionario de listas de adyacencia con costos
graph = {
    'A': [('B', 4), ('C', 2)],
    'B': [('A', 4), ('D', 5), ('E', 1)],
    'C': [('A', 2), ('F', 3), ('G', 7)],
    'D': [('B', 5)],
    'E': [('B', 1), ('H', 8)],
    'F': [('C', 3)],
    'G': [('C', 7)],
    'H': [('E', 8)]
}

# Llamada a la función UCS desde el nodo 'A' hacia el nodo 'G'
print("Recorrido UCS:")
uniform_cost_search(graph, 'A', 'G')

# Este código utiliza una cola de prioridad implementada con un heap (usando heapq) para explorar los nodos de menor a mayor costo.
# La variable priority_queue contiene tuplas (costo acumulado, nodo) que se extraen de forma que se priorizan los nodos con menor costo.

# Ajusta el grafo y los nodos de inicio y objetivo según tus necesidades.
# Ten en cuenta que este es un ejemplo básico, y en aplicaciones prácticas, podrías necesitar manejar más detalles, como la reconstrucción del camino óptimo o la gestión de ciclos.
```
