```python
def dfs(graph, start, visited=None):
    if visited is None:
        visited = set()
    visited.add(start)
    print(start, end=' ')

    for neighbor in graph[start]:
        if neighbor not in visited:
            dfs(graph, neighbor, visited)

# Ejemplo de uso
# Se representa un grafo como un diccionario de listas de adyacencia
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F', 'G'],
    'D': ['B'],
    'E': ['B', 'H'],
    'F': ['C'],
    'G': ['C'],
    'H': ['E']
}

# Llamada a la función DFS desde el nodo 'A'
print("Recorrido DFS:")
dfs(graph, 'A')

# En este ejemplo, el grafo está representado por un diccionario donde las claves son los nodos y los valores son listas de nodos vecinos.
# La función dfs realiza la búsqueda en profundidad, y se inicia desde un nodo de inicio (en este caso, 'A').
# La función imprime los nodos visitados en el orden en que se descubren.

# Puedes ajustar el grafo y el nodo de inicio según tus necesidades.
# Ten en cuenta que esta implementación asume que el grafo es no dirigido y que no hay ciclos infinitos en el grafo.
# Para grafos grandes o complejos, también puede ser necesario implementar manejo de ciclos y otros elementos de control adicionales.
```
