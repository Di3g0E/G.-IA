Acceso a los datos:

```python
# Leer el archivo JSON
with open(r'C:/Users/Diego/OneDrive - Universidad Rey Juan Carlos/Documentos/GIA_URJC/Curso 2023-24/G.-IA/G.-IA/Curso_2/Cuatri_2/AprendizajeAutomatico_1/DataFrames/TumoresCerebrales/train/_annotations.coco.json') as f:
    data = json.load(f)
print(data)

# Extraer la lista de im√°genes
images_data = data['images']

# Crear el DataFrame de im√°genes
images_df = pd.DataFrame(images_data)

# Convertir la lista de anotaciones en un DataFrame
annotations_df = pd.DataFrame(data['annotations'])

# Fusionar los DataFrames basados en la columna 'image_id' y 'id':
df = pd.merge(images_df, annotations_df, left_on='id', right_on='image_id')
```
